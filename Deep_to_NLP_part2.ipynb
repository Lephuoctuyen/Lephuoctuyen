***Sequence-Pair Classification***

Ví dụ, hãy xem xét một tác vụ phân loại mối quan hệ suy diễn với tập dữ liệu Multi-Genre Natural Language Inference (MultiNLI) (Williams và cộng sự, 2018). Trong tác vụ suy luận ngôn ngữ tự nhiên hoặc NLI, còn được gọi là nhận diện suy diễn văn bản, một mô hình sẽ được cung cấp một cặp câu và phải phân loại mối quan hệ giữa ý nghĩa của chúng. Ví dụ, trong tập dữ liệu MultiNLI, các cặp câu được gán một trong ba nhãn: "suy diễn" (entails), "mâu thuẫn" (contradicts), và "trung lập" (neutral). Những nhãn này mô tả mối quan hệ giữa ý nghĩa của câu thứ nhất (câu tiền đề) và ý nghĩa của câu thứ hai (câu giả thuyết).

Nội dung trên mô tả cách thức điều chỉnh một mô hình để giải quyết các tác vụ phân loại cặp chuỗi đầu vào, đặc biệt là trong lĩnh vực suy luận ngôn ngữ tự nhiên (Natural Language Inference - NLI) với tập dữ liệu Multi-Genre Natural Language Inference (MultiNLI) từ Williams và cộng sự (2018). Trong NLI, mô hình được cung cấp một cặp câu và phải phân loại mối quan hệ giữa ý nghĩa của chúng thành một trong ba nhãn: **suy diễn** (entails), **mâu thuẫn** (contradicts) và **trung lập** (neutral).


1. **Phân loại cặp chuỗi**: Các ứng dụng thực tế thuộc loại này gồm phát hiện diễn giải (hai câu có cùng nghĩa không?), suy diễn logic (câu A có suy diễn logic câu B không?) và tính liền mạch của văn bản (câu B có tiếp nối hợp lý câu A không?).

2. **Quá trình fine-tuning**: Mô hình sử dụng phương pháp fine-tuning tương tự như huấn luyện với mục tiêu NSP (Next Sentence Prediction) để xử lý các cặp câu có nhãn. Các cặp câu được tách nhau bằng ký tự `[SEP]`, còn ký tự `[CLS]` được sử dụng để biểu diễn toàn bộ ý nghĩa của cặp câu khi đi qua các lớp mạng, tạo ra vector đầu ra phục vụ cho tác vụ phân loại. Vector này được nhân với các trọng số của lớp phân loại và đưa qua hàm softmax để dự đoán nhãn.

3. **Ví dụ minh họa từ MultiNLI**:
   - **Trung lập**: "Jon walked back to the town to the smithy." và "Jon traveled back to his hometown."
   - **Mâu thuẫn**: "Tourist Information offices can be very helpful." và "Tourist Information offices are never of any help."
   - **Suy diễn**: "I’m confused." và "Not all of it is very clear to me."

4. **Nhãn**: "Mâu thuẫn" nghĩa là câu tiền đề mâu thuẫn với câu giả thuyết; "suy diễn" nghĩa là câu tiền đề suy diễn câu giả thuyết; "trung lập" nghĩa là không có mối quan hệ suy diễn hay mâu thuẫn rõ ràng. Mô hình sử dụng vector `[CLS]` để huấn luyện bộ phân loại ba chiều dựa trên tập huấn luyện của MultiNLI.

### Fine-Tuning for Sequence Labelling: Named Entity Recognition

image.png

 ***Named entity recognition***.

Nội dung mô tả về nhiệm vụ nhận dạng thực thể có tên (Named Entity Recognition - NER), là quá trình tìm các đoạn văn bản chứa tên riêng và gắn nhãn loại của thực thể đó. Các loại thực thể thường gặp gồm:

- **PER** (người), **LOC** (địa điểm), **ORG** (tổ chức), và **GPE** (thực thể địa-chính trị).
- Các thực thể còn có thể bao gồm ngày, thời gian, hoặc các biểu thức số học như giá tiền.

Ví dụ minh họa cho NER gồm các thực thể như "United Airlines" (tổ chức), "Friday" (thời gian), và "$6" (tiền tệ). Những nhãn này được sử dụng trong nhiều tác vụ xử lý ngôn ngữ tự nhiên như liên kết văn bản với thông tin trong các nguồn tri thức, phân tích thái độ đối với các thực thể, hoặc ẩn danh hóa văn bản để bảo vệ quyền riêng tư.

 Các thách thức trong NER:
1. **Phân đoạn thực thể**: Khó xác định đoạn nào là thực thể, vì đa số từ trong văn bản không phải thực thể.
2. **Sự mơ hồ về loại thực thể**: Một từ có thể mang nhiều nghĩa khác nhau, chẳng hạn "Washington" có thể ám chỉ một người (PER), đội thể thao (ORG), thành phố (LOC), hoặc chính phủ Mỹ (GPE).

Các ví dụ về sự mơ hồ của từ "Washington" cho thấy mức độ phức tạp khi gán nhãn đúng loại thực thể trong ngữ cảnh.

 ***BIO Tagging***

BIO tagging là một phương pháp tiêu chuẩn trong gán nhãn chuỗi cho các bài toán nhận dạng chuỗi như NER (Nhận dạng thực thể có tên), được giới thiệu bởi Ramshaw và Marcus vào năm 1995. Phương pháp này cho phép xử lý NER như một nhiệm vụ gán nhãn từ từng từ, với các nhãn giúp xác định cả ranh giới và loại của thực thể.

Ví dụ, xét câu:
"[PER Jane Villanueva] of [ORG United], a unit of [ORG United Airlines Holding], said the fare applies to the [LOC Chicago] route."

BIO tagging gán nhãn như sau:
- Nhãn B (Begin) cho các từ bắt đầu một chuỗi thực thể.
- Nhãn I (Inside) cho các từ nằm trong chuỗi.
- Nhãn O (Outside) cho các từ không thuộc bất kỳ chuỗi thực thể nào.

Với BIO tagging, mỗi loại thực thể có nhãn B và I riêng, ngoài nhãn O duy nhất. Số lượng nhãn cần thiết là \(2n+1\), trong đó \(n\) là số loại thực thể. BIO có thể biểu diễn thông tin giống như ký hiệu đóng ngoặc, nhưng cho phép mô hình hóa bài toán một cách đơn giản như gán nhãn từ loại (POS tagging).

Có các biến thể của BIO tagging:
- **IO tagging**: Bỏ nhãn B, chỉ sử dụng nhãn I và O, làm mất thông tin về điểm bắt đầu.
- **BIOES tagging**: Thêm nhãn E (End) cho từ cuối trong chuỗi và S (Single) cho chuỗi chỉ có một từ.

Hình 11.12 minh họa cách sử dụng các phương pháp IO, BIO, và BIOES cho gán nhãn chuỗi trong NER.

***Sequence Labeling***

Trong gán nhãn chuỗi, chúng ta truyền vector đầu ra cuối cùng tương ứng với mỗi token đầu vào đến một bộ phân loại, bộ phân loại này sẽ tạo ra phân phối softmax trên tập hợp các nhãn có thể có. Đối với một bộ phân loại với một lớp truyền thẳng duy nhất, tập hợp trọng số cần học là $ W_K $ với kích thước $[d \times k]$, trong đó $ k $ là số lượng nhãn có thể có cho tác vụ.

Một cách tiếp cận tham lam (greedy), khi chọn nhãn có xác suất lớn nhất (argmax) cho mỗi token, có thể được sử dụng để tạo ra chuỗi nhãn đầu ra cuối cùng. Hình 11.13 minh họa ví dụ cho phương pháp này, trong đó $ y_i $ là vector xác suất trên các nhãn, và $ k $ là chỉ số của các nhãn.

$$
y_i = \text{softmax}(h_L^i W_K)
$$

$$
t_i = \arg \max_k(y_i)
$$

Ngoài ra, phân phối xác suất trên các nhãn do softmax cung cấp cho mỗi token đầu vào cũng có thể được truyền đến một lớp điều kiện ngẫu nhiên trường (Conditional Random Field - CRF), cho phép tính đến các chuyển tiếp toàn cục giữa các nhãn (xem Chương 17 về CRFs).

***Tokenization and NER***

**Tokenization và Nhận dạng thực thể có tên (NER)**

Dữ liệu huấn luyện có giám sát cho bài toán NER thường dưới dạng các nhãn BIO đi kèm với văn bản đã được phân đoạn theo từ. Ví dụ, câu chứa hai thực thể có tên sau:

```
[LOC Mt. Sanitas] is in [LOC Sunshine Canyon].
```

sẽ có bộ nhãn BIO cho từng từ như sau:

```
Mt.         B-LOC
Sanitas     I-LOC
is          O
in          O
Sunshine    B-LOC
Canyon      I-LOC
.           O
```

**Tuy nhiên, dãy token WordPiece**của câu này không khớp trực tiếp với nhãn BIO trong phần chú thích:

```
Mt, ., San, ##itas, is, in, Sunshine, Canyon, .
```

Để xử lý sự không khớp này, chúng ta cần một cách để gán nhãn BIO cho các token con (subword tokens) trong quá trình huấn luyện và một cách tương ứng để lấy lại các nhãn cấp từ từ các token con trong quá trình giải mã. Trong quá trình huấn luyện, chúng ta có thể gán nhãn chuẩn (gold-standard tag) liên quan đến mỗi từ cho tất cả các token con xuất phát từ từ đó.

Trong quá trình giải mã, cách tiếp cận đơn giản nhất là sử dụng nhãn BIO với xác suất cao nhất (argmax) được gán cho token con đầu tiên của một từ. Ví dụ, nhãn BIO gán cho “Mt” sẽ được gán cho “Mt.”, và nhãn gán cho “San” sẽ được gán cho “Sanitas”, bỏ qua thông tin trong các nhãn gán cho “.” và “##itas”. Các phương pháp phức tạp hơn kết hợp phân phối xác suất của các nhãn trên các token con để cố gắng tìm nhãn tối ưu ở cấp từ.``

***Evaluating Named Entity Recognition***


Các hệ thống NER được đánh giá dựa trên các chỉ số: Recall, Precision và F1. Cụ thể:

- **Recall** là tỷ lệ giữa số lượng phản hồi được gán nhãn đúng với tổng số phản hồi lẽ ra phải được gán nhãn.
- **Precision** là tỷ lệ giữa số lượng phản hồi được gán nhãn đúng với tổng số phản hồi đã được gán nhãn.
- **F1-measure** là trung bình điều hòa giữa Precision và Recall.

Để biết liệu sự khác biệt giữa điểm F1 của hai hệ thống NER có thực sự ý nghĩa hay không, ta sử dụng kiểm định bootstrap cặp (paired bootstrap test) hoặc kiểm định ngẫu nhiên tương tự (randomization test) (xem Mục 4.9).

Trong bài toán gán nhãn thực thể, đơn vị phản hồi là **thực thể** chứ không phải là từ. Ví dụ, trong hình 11.12, các thực thể “Jane Villanueva” và “United Airlines Holding”, cùng với phần không phải là thực thể “discussed” sẽ được tính là ba phản hồi riêng biệt.

Việc gán nhãn thực thể có tên bao gồm một thành phần phân đoạn, điều này không có trong các tác vụ như phân loại văn bản hay gán nhãn loại từ (part-of-speech tagging), gây ra một số vấn đề trong việc đánh giá. Ví dụ, nếu một hệ thống chỉ gán nhãn “Jane” mà không gán nhãn “Jane Villanueva” là một thực thể người, nó sẽ tạo ra hai lỗi: một dương tính sai (false positive) cho nhãn “O” và một âm tính sai (false negative) cho nhãn “I-PER”. Ngoài ra, việc sử dụng thực thể làm đơn vị phản hồi nhưng từ lại là đơn vị huấn luyện tạo ra sự không đồng nhất giữa điều kiện huấn luyện và thử nghiệm.

### Summary

**Chương này giới thiệu về bộ mã hóa hai chiều (bidirectional encoder) và mô hình ngôn ngữ mặt nạ (masked language model)**. Dưới đây là tóm tắt các điểm chính đã được đề cập:

- **Bộ mã hóa hai chiều** có thể được sử dụng để tạo ra các biểu diễn ngữ cảnh hóa của các embedding đầu vào bằng cách sử dụng toàn bộ ngữ cảnh đầu vào.
  
- **Các mô hình ngôn ngữ tiền huấn luyện dựa trên bộ mã hóa hai chiều** có thể được học bằng cách sử dụng mục tiêu của mô hình ngôn ngữ mặt nạ, trong đó mô hình được huấn luyện để dự đoán thông tin bị thiếu trong đầu vào.

- **Đầu ra của mỗi khối hoặc thành phần transformer** trong một cột token cụ thể là một embedding ngữ cảnh, biểu diễn một khía cạnh nào đó của ý nghĩa của token trong ngữ cảnh.

- **Nghĩa của từ** là một biểu diễn rời rạc của một khía cạnh ý nghĩa của từ. Embedding ngữ cảnh cung cấp một mô hình ý nghĩa liên tục có chiều cao, phong phú hơn so với các nghĩa hoàn toàn rời rạc.

- **Cosine giữa các embedding ngữ cảnh** có thể được sử dụng như một cách để mô hình hóa sự tương đồng giữa hai từ trong ngữ cảnh, mặc dù trước tiên cần có một số biến đổi trên các embedding này.

- **Các mô hình ngôn ngữ tiền huấn luyện** có thể được tinh chỉnh cho các ứng dụng cụ thể bằng cách thêm các lớp phân loại nhẹ phía trên đầu ra của mô hình tiền huấn luyện.

- **Các ứng dụng này** có thể bao gồm các tác vụ phân loại chuỗi như phân tích cảm xúc, phân loại cặp chuỗi như suy luận ngôn ngữ tự nhiên, hoặc gán nhãn chuỗi như nhận dạng thực thể có tên (NER).

## Model Alignment, Prompting,and In-Context Learning

### Prompting

**A prompt is a text string that a user issues to a language model to get the model to do something useful.**

In prompting, the user’s prompt string is passed to the language model, which **iteratively generates tokens conditioned on the prompt**. Thus the prompt creates a context that guides LLMs to generate useful outputs to **achieve some user goal**. The process of **finding effective prompts** for a task is known as **prompt engineering**

image.png

Cách thiết kế *prompt* để hướng dẫn LLM thực hiện đúng nhiệm vụ. Mỗi *template* gồm phần *input* và *prompt* để tạo ra *filled prompts* phù hợp cho mô hình. Một cấu trúc phổ biến là đặt hướng dẫn ở cuối *prompt*, giúp giới hạn khả năng sinh đầu ra. Ví dụ, nếu không thiết kế rõ ràng, mô hình có thể hiểu sai yêu cầu (như dịch ngôn ngữ) và tạo nội dung không mong muốn. Để tăng độ chính xác, *prompt* có thể chỉ định sẵn các câu trả lời khả dĩ, giúp mô hình thực hiện đúng nhiệm vụ.

xem thêm thiết kế một cấu trúc cho prompts rõ ràng : https://arxiv.org/pdf/2102.07350

**Với Input (Context) + prompt theo sau.**

Một kiểu *prompt* giới hạn hơn có thể chỉ định trước các câu trả lời khả dĩ trong chính *prompt*. Ví dụ, dưới đây là một template *prompt* cho phân tích cảm xúc, trong đó xác định trước các câu trả lời tiềm năng:

Một *prompt* bao gồm đánh giá và câu chưa hoàn chỉnh:
input : ....

- **Human:** Prompt : Bạn nghĩ rằng “*input*” có cảm xúc tích cực hay tiêu cực?
- **Lựa chọn:**
  - (P) Tích cực
  - (N) Tiêu cực
- **Assistant:** Tôi tin rằng câu trả lời đúng nhất là: (N) or (P)

image.png

Đoạn văn này giải thích cách tạo ra các *prompt* tinh vi hơn cho mô hình ngôn ngữ lớn (LLM) và quy trình để chuyển đổi nhiệm vụ thành dạng phù hợp cho LLM.

1. **Các đặc điểm của *prompt***: Prompt này chỉ định rõ hai lựa chọn cho câu trả lời (P) và (N), và kết thúc bằng dấu ngoặc mở, điều này mạnh mẽ gợi ý rằng câu trả lời sẽ là (P) hoặc (N). Nó cũng xác định vai trò của mô hình ngôn ngữ là một trợ lý.

2. **Khả năng mở rộng của *prompt***: Chúng ta có thể phát triển hơn nữa các *prompt*, ví dụ như yêu cầu tóm tắt phải có độ dài nhất định, hoặc yêu cầu câu trả lời theo một phong cách hoặc vai trò nhất định. Chúng ta cũng có thể chỉ định đầu ra có cấu trúc hơn, sử dụng ngôn ngữ lập trình hoặc định dạng trao đổi dữ liệu như JSON. Ngoài ra, có thể yêu cầu mô hình phân tách các tác vụ phức tạp, sử dụng các phương pháp như *chain-of-thought*.

3. **Quy trình tạo *prompt***:
   - Đầu tiên, cho một nhiệm vụ nhất định, phát triển một mẫu *prompt* cụ thể cho nhiệm vụ đó, với một tham số tự do cho văn bản đầu vào.
   - Sau đó, sử dụng đầu vào và mẫu cụ thể đó để tạo ra một *prompt* đã điền, sau đó gửi nó tới mô hình ngôn ngữ đã được huấn luyện trước.
   - Tiếp theo, sử dụng phương pháp *autoregressive decoding* để tạo ra một chuỗi đầu ra token.
   - Cuối cùng, đầu ra của mô hình có thể được sử dụng trực tiếp như đầu ra mong muốn (trong các tác vụ tự nhiên như dịch thuật hoặc tóm tắt), hoặc có thể trích xuất câu trả lời phù hợp với nhiệm vụ từ đầu ra đã tạo (như trong trường hợp phân loại).

Các hướng dẫn và quy trình này giúp tối ưu hóa khả năng của LLM trong việc hoàn thành các nhiệm vụ cụ thể.

 ***Learning from Demonstrations: Few-Shot Promptin***

Thường thì có thể cải thiện một *prompt* bằng cách bao gồm một số ví dụ đã được gán nhãn trong mẫu *prompt*. Chúng tôi gọi những ví dụ này là *demonstrations*. Nhiệm vụ sử dụng ví dụ trong việc tạo *prompt* đôi khi được gọi là *few-shot prompting*, tương phản với *zero-shot prompting*, có nghĩa là các hướng dẫn không bao gồm ví dụ đã được gán nhãn. Hình 12.2 minh họa một ví dụ *few-shot* từ một nhiệm vụ hỏi đáp trích xuất. Ngữ cảnh kết hợp giữa định nghĩa nhiệm vụ cùng với ba cặp câu hỏi và câu trả lời tiêu chuẩn từ tập huấn luyện.

**Số Lượng Minh Chứng**

Số lượng minh chứng không cần phải lớn. Một số lượng nhỏ các ví dụ đã được chọn ngẫu nhiên có thể đủ để cải thiện hiệu suất so với cài đặt zero-shot. Thực tế, sự cải thiện lớn nhất trong *few-shot prompting* thường đến từ ví dụ huấn luyện đầu tiên, với những lợi ích giảm dần cho các minh chứng tiếp theo. Điều này trái ngược với việc tinh chỉnh các đầu phân loại chuyên biệt mà chúng ta đã thấy trong Chương 11, nơi mà việc có nhiều ví dụ là hữu ích.

**Tại sao không nên có quá nhiều minh chứng (Demonstrations)?** Lý do là lợi ích chính từ các ví dụ là để minh họa nhiệm vụ cần thực hiện cho LLM và định dạng của chuỗi, chứ không phải để cung cấp thông tin liên quan đến câu trả lời đúng cho bất kỳ câu hỏi cụ thể nào. Thực tế, các minh chứng có câu trả lời sai vẫn có thể cải thiện hệ thống (Min et al., 2022; Webson và Pavlick, 2022). Thêm quá nhiều ví dụ dường như khiến mô hình bị *overfit* vào chi tiết của các ví dụ cụ thể đã chọn và kém tổng quát hơn.

**Cách Chọn Minh Chứng(Demonstrations)?** Các minh chứng thường được tạo ra bằng cách định dạng các ví dụ được lấy từ tập huấn luyện có nhãn. Có một số quy tắc về điều gì tạo nên một minh chứng tốt. Ví dụ, việc sử dụng các minh chứng tương tự với đầu vào hiện tại dường như cải thiện hiệu suất. Do đó, có thể hữu ích khi tìm kiếm động các minh chứng cho từng đầu vào, dựa trên sự tương tự với ví dụ hiện tại (ví dụ, so sánh nhúng của ví dụ hiện tại với nhúng của từng ví dụ trong tập huấn luyện để tìm ra top-T tốt nhất).

Nhưng nói chung, cách tốt nhất để chọn minh chứng từ tập huấn luyện là một cách có tính toán: chọn tập hợp các minh chứng mà tăng cường hiệu suất nhiệm vụ nhất cho *prompt* trên tập kiểm tra. Hiệu suất nhiệm vụ cho phân tích cảm xúc hoặc hỏi đáp lựa chọn có thể được đo bằng độ chính xác; đối với dịch máy có thể đo bằng chrF, và đối với tóm tắt qua Rouge. Các hệ thống như DSPy (Khat tab et al., 2024), một khung để tối ưu hóa thuật toán cho LM *prompts*, có thể tự động tìm ra tập hợp minh chứng tối ưu để bao gồm bằng cách tìm kiếm qua không gian các minh chứng có thể bao gồm. Chúng ta sẽ trở lại với việc tối ưu hóa *prompt* tự động trong Mục 12.5.

 ***In-Context Learning and Induction Heads***



Để khiến một mô hình thực hiện đúng điều mà chúng ta mong muốn, việc nhắc nhở (prompting) hoàn toàn khác biệt so với tiền huấn luyện (pretraining). **Học thông qua tiền huấn luyện có nghĩa là cập nhật các tham số của mô hình bằng cách sử dụng phương pháp hạ gradient dựa theo một hàm mất mát** nào đó. **Nhưng nhắc nhở với các ví dụ có thể dạy cho mô hình thực hiện một tác vụ mới**. Mô hình đang học được điều gì đó trong quá trình xử lý lời nhắc.

Ngay cả khi không có các ví dụ, chúng ta vẫn có thể xem quá trình nhắc nhở như một dạng học hỏi. Ví dụ, càng đi sâu vào lời nhắc, mô hình càng có xu hướng đoán tốt hơn các token tiếp theo. Thông tin trong ngữ cảnh đang giúp mô hình tăng khả năng dự đoán.

**Học trong ngữ cảnh**  
Chúng tôi dùng thuật ngữ “học trong ngữ cảnh” (in-context learning) để chỉ loại hình học hỏi mà các mô hình ngôn ngữ thực hiện từ lời nhắc. Học trong ngữ cảnh có nghĩa là mô hình ngôn ngữ học để làm các nhiệm vụ mới, dự đoán tốt hơn các token, hoặc nói chung là giảm mức mất mát, nhưng **không cần bất kỳ cập nhật tham số nào dựa trên gradient cho mô hình**.

Học trong ngữ cảnh hoạt động như thế nào? Mặc dù chúng ta **chưa biết chắc chắn**, có một số ý tưởng thú vị. Một giả thuyết dựa trên ý tưởng về các **đầu cảm ứng** (induction heads) (Elhage et al., 2021; Olsson et al., 2022). Các đầu cảm ứng là một loại mạch, một thành phần trừu tượng của mạng lưới. Mạch đầu cảm ứng là một phần của tính toán sự chú ý trong transformers, được phát hiện khi nghiên cứu các mô hình ngôn ngữ nhỏ chỉ có 1-2 đầu chú ý.

**Chức năng của đầu cảm ứng là dự đoán các chuỗi lặp lại**. **Ví dụ, nếu nó thấy mẫu AB...A trong một chuỗi đầu vào, nó sẽ dự đoán rằng B sẽ theo sau, theo quy tắc hoàn thành mẫu AB...A B.** Nó thực hiện điều này bằng cách có một thành phần khớp tiền tố trong tính toán sự chú ý, khi nhìn vào token hiện tại A, nó tìm kiếm lại trong ngữ cảnh để tìm một lần xuất hiện trước đó của A. Nếu tìm thấy, đầu cảm ứng có một cơ chế sao chép để "sao chép" token B theo sau A trước đó, bằng cách tăng xác suất B sẽ xuất hiện tiếp theo. Hình 12.3 cho thấy một ví dụ minh họa quá trình này.

image.png

Để xác định các đầu cảm ứng trong các mô hình, Olsson et al. (2022) đề xuất rằng một phiên bản tổng quát mờ của quy tắc hoàn thành mẫu này, thực hiện quy tắc dạng A * B * ...A B*, nơi A* ≈ A và B* ≈ B (theo một cách nào đó chúng có sự tương đồng về ngữ nghĩa), có thể là yếu tố chịu trách nhiệm cho việc học trong ngữ cảnh. Bằng chứng gợi ý cho giả thuyết này đến từ nghiên cứu của Crosbie và Shutova (2022), cho thấy rằng việc loại bỏ đầu cảm ứng sẽ làm giảm khả năng học trong ngữ cảnh của mô hình một cách rõ rệt.

**Loại bỏ (ablation)** là một thuật ngữ y học có nghĩa là loại bỏ một cái gì đó. Trong nghiên cứu giải thích NLP, chúng tôi dùng nó như một công cụ để kiểm tra các hiệu ứng nhân quả; nếu chúng ta loại bỏ một nguyên nhân được giả định, ta mong rằng hiệu ứng sẽ biến mất. Crosbie và Shutova (2022) loại bỏ đầu cảm ứng bằng cách tìm các đầu chú ý hoạt động như đầu cảm ứng trên các chuỗi đầu vào ngẫu nhiên, sau đó đặt các phần tử trong ma trận đầu ra $ W_O $ bằng 0 để làm mất tác dụng của các đầu này. Thật vậy, họ nhận thấy rằng các mô hình bị loại bỏ có hiệu suất học trong ngữ cảnh kém hơn nhiều: chúng gặp khó khăn đáng kể khi học từ các ví dụ trong lời nhắc.

**Kết quả là các mô hình sau khi bị loại bỏ đầu cảm ứng có hiệu suất học kém hơn hẳn trong việc học từ các ví dụ trong lời nhắc, chứng tỏ rằng các đầu cảm ứng này đóng vai trò quan trọng trong khả năng học của mô hình khi chỉ dựa vào ngữ cảnh (mà không có bất kỳ cập nhật tham số nào).**

**4. Kỹ thuật tạo Prompt**  
Kỹ thuật tạo prompt là quá trình tạo ra hàm prompt \( f_{prompt}(x) \) nhằm mang lại hiệu suất tối ưu nhất cho tác vụ kế tiếp. Trong nhiều công trình trước đây, quá trình này bao gồm việc xây dựng mẫu prompt (prompt template) thủ công, trong đó một kỹ sư hoặc thuật toán tìm kiếm mẫu tốt nhất cho mỗi tác vụ mà mô hình được kỳ vọng thực hiện. Như trong mục “Kỹ thuật tạo Prompt” của Hình 1, ta cần cân nhắc hình dạng của prompt, sau đó quyết định thực hiện thủ công hay tự động để tạo ra các prompt với hình dạng mong muốn, chi tiết như sau.

**4.1 Hình dạng của Prompt**  
Có hai loại prompt chính: prompt điền vào chỗ trống (cloze prompts) (Petroni và cộng sự, 2019; Cui và cộng sự, 2021), điền vào các khoảng trống trong một chuỗi văn bản; và prompt tiền tố (prefix prompts) (Li và Liang, 2021; Lester và cộng sự, 2021), tiếp tục một chuỗi tiền tố. Lựa chọn loại prompt tùy thuộc vào tác vụ và mô hình được sử dụng. Đối với các tác vụ sinh dữ liệu hoặc giải quyết bằng mô hình tự hồi quy, prompt tiền tố thường phù hợp hơn do phù hợp với mô hình từ trái sang phải. Đối với các tác vụ giải quyết bằng mô hình bị che (masked LM), prompt điền vào chỗ trống sẽ hiệu quả hơn vì phù hợp với dạng bài toán huấn luyện trước. Cuối cùng, với các bài toán cần nhiều đầu vào như phân loại cặp văn bản, mẫu prompt phải chứa đủ chỗ cho hai đầu vào, như [X1] và [X2].

**4.2 Tạo Mẫu Thủ Công**  
Cách tự nhiên nhất để tạo prompt là tạo mẫu thủ công dựa trên trực giác của con người. Ví dụ, tập dữ liệu LAMA (Petroni và cộng sự, 2019) cung cấp các mẫu cloze được tạo thủ công để kiểm tra kiến thức trong các mô hình ngôn ngữ (LM). Brown và cộng sự (2020) tạo các prompt tiền tố thủ công để xử lý nhiều tác vụ như trả lời câu hỏi, dịch, và kiểm tra hiểu biết thông thường. Schick và Schütze (2020, 2021a,b) sử dụng mẫu định sẵn trong bối cảnh học vài mẫu cho các tác vụ phân loại văn bản và sinh văn bản có điều kiện.

**4.3 Tự Động Hóa Học Mẫu**  
Mặc dù tạo mẫu thủ công là trực quan và cho phép giải quyết nhiều tác vụ với độ chính xác nhất định, nhưng phương pháp này cũng có hạn chế: (1) tạo và thử nghiệm các prompt đòi hỏi thời gian và kinh nghiệm, đặc biệt đối với các tác vụ phức tạp như phân tích ngữ nghĩa (Shin và cộng sự, 2021); (2) ngay cả những người thiết kế prompt có kinh nghiệm cũng có thể không tìm ra prompt tối ưu. Để giải quyết các vấn đề này, một số phương pháp đã được đề xuất để tự động hóa quy trình thiết kế mẫu. Đặc biệt, các prompt tự động có thể chia thành prompt rời rạc (discrete prompts) và prompt liên tục (continuous prompts).

**4.3.1 Prompt Rời Rạc (Discrete Prompts)**  
Các nghiên cứu về prompt rời rạc (hay còn gọi là hard prompts) tự động tìm kiếm các mẫu trong không gian rời rạc, thường là các cụm từ trong ngôn ngữ tự nhiên. Một số phương pháp gồm:

- **D1: Khai thác Prompt (Prompt Mining)**  
  Phương pháp MINE của Jiang và cộng sự (2020c) là phương pháp khai thác để tìm các mẫu dựa trên tập dữ liệu đầu vào và đầu ra. Phương pháp này khai thác một tập dữ liệu lớn (ví dụ Wikipedia) để tìm các chuỗi có chứa đầu vào và đầu ra, sau đó tìm các từ ở giữa hoặc các đường dẫn phụ thuộc giữa các đầu vào và đầu ra để tạo mẫu.
  
- **D2: Diễn giải Prompt (Prompt Paraphrasing)**  
  Các phương pháp diễn giải lấy một prompt ban đầu (ví dụ: prompt thủ công hoặc khai thác được) và diễn giải thành các prompt khác, sau đó chọn ra prompt có độ chính xác cao nhất cho tác vụ. Việc diễn giải có thể thực hiện bằng nhiều cách, ví dụ dịch vòng qua ngôn ngữ khác, sử dụng từ đồng nghĩa, hoặc sử dụng mạng nơ-ron tối ưu hóa cho prompt.

- **D3: Tìm kiếm Dựa trên Gradient (Gradient-based Search)**  
  Wallace và cộng sự (2019a) sử dụng tìm kiếm dựa trên gradient qua các token để tìm chuỗi ngắn có thể kích hoạt LM sinh ra dự đoán mong muốn. Shin và cộng sự (2020) cũng sử dụng phương pháp này để tự động tìm các token mẫu và đạt hiệu quả cao trong các tình huống prompting.

- **D4: Sinh Prompt (Prompt Generation)**  
  Một số nghiên cứu xem việc tạo prompt là một tác vụ sinh văn bản và sử dụng các mô hình ngôn ngữ tự nhiên cho nhiệm vụ này. Ví dụ, Gao và cộng sự (2021) sử dụng mô hình T5 để sinh các token mẫu và các từ khóa đặc trưng cho từng miền (DRF) để thêm vào đầu vào và phục vụ cho các tác vụ sau.

- **D5: Đánh giá Prompt (Prompt Scoring)**  
  Davison và cộng sự (2019) nghiên cứu hoàn thiện cơ sở tri thức và thiết kế mẫu cho các input (bộ ba chủ ngữ-quan hệ-vị ngữ) bằng các mô hình ngôn ngữ.

 **4.3.2 Các Prompt Liên Tục**

Vì mục tiêu của việc xây dựng prompt là để giúp mô hình ngôn ngữ (LM) thực hiện tốt nhiệm vụ, không cần phải giới hạn prompt trong ngôn ngữ tự nhiên dễ hiểu. Vì vậy, có các phương pháp tạo prompt liên tục (hay còn gọi là **soft prompt**) trực tiếp trong không gian embedding của mô hình. Các prompt liên tục bỏ qua hai ràng buộc sau: (1) không yêu cầu các embedding của từ trong template phải là từ trong ngôn ngữ tự nhiên (ví dụ như tiếng Anh), và (2) không yêu cầu template phải được tham số hóa dựa trên các tham số của mô hình LM đã huấn luyện. Thay vào đó, các template có tham số riêng có thể được tinh chỉnh dựa trên dữ liệu huấn luyện cho nhiệm vụ cụ thể. Dưới đây là một số phương pháp tiêu biểu:

***C1: Tuning Tiền Tố (Prefix Tuning)***

**Prefix Tuning** (Li và Liang, 2021) là phương pháp thêm một chuỗi các vector liên tục đặc thù cho nhiệm vụ vào trước chuỗi đầu vào, đồng thời giữ nguyên các tham số của mô hình ngôn ngữ (LM). Về mặt toán học, phương pháp này tối ưu hóa mục tiêu log-likelihood bằng cách sử dụng một ma trận tiền tố có thể huấn luyện được $ M $ và một mô hình LM đã được huấn luyện sẵn và cố định.

image.png

Trong phương trình (2), $ h_{<i} = [h^{(1)}_{<i}; \dots; h^{(n)}_{<i}] $ là sự kết hợp của tất cả các lớp mạng neural tại thời điểm $ i $. Giá trị này được lấy trực tiếp từ $ M $ nếu thời điểm tương ứng nằm trong tiền tố (tức là $ h_i = M[i] $); nếu không, nó được tính toán từ mô hình LM đã được huấn luyện.

Qua thí nghiệm, Li và Liang (2021) nhận thấy rằng học dựa trên tiền tố liên tục nhạy cảm hơn với các khởi tạo khác nhau trong điều kiện dữ liệu ít so với việc sử dụng prompt rời rạc với các từ thật. Tương tự, Lester và cộng sự (2021) thêm các token đặc biệt vào trước chuỗi đầu vào để tạo template và tinh chỉnh embedding của các token này trực tiếp. Phương pháp này so với Li và Liang (2021) ít bổ sung tham số hơn vì không thêm các tham số có thể tinh chỉnh trong mỗi lớp mạng.

Tsimpoukelli và cộng sự (2021) huấn luyện một bộ mã hóa hình ảnh để mã hóa một hình ảnh thành chuỗi embedding, sau đó có thể sử dụng làm prompt cho LM tự hồi quy đã đóng băng để tạo ra chú thích thích hợp. Họ cho thấy rằng mô hình thu được có thể thực hiện học ít dữ liệu (few-shot learning) cho các nhiệm vụ liên quan đến ngôn ngữ-hình ảnh, chẳng hạn như trả lời câu hỏi dựa trên hình ảnh. Khác với hai phương pháp trên, tiền tố được sử dụng trong nghiên cứu của Tsimpoukelli và cộng sự (2021) phụ thuộc vào mẫu, cụ thể là một biểu diễn của các hình ảnh đầu vào, thay vì embedding cho nhiệm vụ.

- **C2: Tuning với Prompt Rời Rạc Khởi Tạo** – Phương pháp này bắt đầu với prompt đã tạo hoặc tìm kiếm qua các phương pháp prompt rời rạc. Zhong và cộng sự (2021b) trước tiên định nghĩa template bằng cách tìm kiếm rời rạc như AUTOPROMPT, sau đó khởi tạo các token ảo dựa trên prompt này và tinh chỉnh embedding để tăng độ chính xác. Qin và Eisner (2021) học một hỗn hợp các template mềm cho mỗi input và trọng số của mỗi template được học từ dữ liệu huấn luyện.

- **C3: Tuning Kết Hợp Prompt Cứng và Mềm (Hard-Soft Prompt Hybrid Tuning)** – Thay vì chỉ dùng prompt có thể học, phương pháp này chèn các embedding có thể học vào trong template hard prompt. Liu và cộng sự (2021b) đề xuất “P-tuning”, trong đó prompt liên tục được học bằng cách chèn các biến có thể huấn luyện vào input đã được embedding. P-tuning cũng sử dụng các token neo liên quan đến nhiệm vụ (ví dụ, “thủ đô” trong bài toán trích xuất quan hệ), các token này không được tinh chỉnh trong quá trình huấn luyện.

### Post-training and Model Alignment

Những hạn chế và thách thức của các mô hình ngôn ngữ lớn (LLM), đặc biệt là khi chúng được sử dụng mà không qua giai đoạn tinh chỉnh. Những mô hình này, được huấn luyện để dự đoán từ tiếp theo, thường không tuân thủ chính xác các chỉ dẫn cụ thể và có thể tạo ra nội dung không liên quan hoặc gây hại. Ví dụ, trong một số trường hợp, mô hình bỏ qua yêu cầu của người dùng và chỉ tạo ra các đoạn văn tiếp nối phù hợp với ngữ cảnh chứ không phải là phản hồi trực tiếp theo yêu cầu.

Nguyên nhân của những vấn đề này là do mục tiêu huấn luyện của mô hình, chỉ đơn thuần là dự đoán từ, không hoàn toàn phù hợp với nhu cầu của con người về sự hữu ích và an toàn trong tương tác.

Để cải thiện khả năng hữu ích và an toàn của LLMs, có hai kỹ thuật thường được áp dụng sau giai đoạn tiền huấn luyện:
1. **Tinh chỉnh theo hướng dẫn (Instruction Tuning hoặc Fine-Tuning có giám sát - SFT)**: Mô hình được tinh chỉnh với một tập hợp các câu hỏi và câu trả lời để cải thiện khả năng tuân thủ chỉ dẫn.
2. **Căn chỉnh theo sở thích (Preference Alignment hoặc Học Tăng cường từ Phản hồi của Con người - RLHF)**: Một mô hình riêng được huấn luyện để đánh giá mức độ đáp ứng của phản hồi với sở thích của con người, từ đó điều chỉnh thêm cho LLM.

Những phương pháp này giúp điều chỉnh các mô hình tiền huấn luyện để tạo ra “mô hình đã căn chỉnh”, phù hợp hơn với kỳ vọng của con người.

### Model Alignment: Instruction Tuning

image.png

Nội dung giới thiệu về kỹ thuật **instruction tuning** (còn gọi là **instruct tuning** hoặc **supervised fine-tuning – SFT**), một phương pháp để cải thiện khả năng tuân theo chỉ dẫn của các mô hình ngôn ngữ lớn (LLMs). Cách này gồm việc tiếp tục huấn luyện một mô hình đã tiền huấn luyện bằng cách dùng tập dữ liệu gồm các câu hỏi và câu trả lời. Quá trình này giúp mô hình không chỉ học các nhiệm vụ cụ thể mà còn nâng cao khả năng làm theo chỉ dẫn nói chung, coi như một dạng học meta.

Instruction tuning là một hình thức học có giám sát, tiếp tục sử dụng mục tiêu dự đoán từ tiếp theo. Tuy nhiên, không giống như trong tiền huấn luyện tự giám sát, ở đây mỗi chỉ dẫn hoặc câu hỏi đều có câu trả lời đúng hoặc phản hồi cụ thể.

So sánh với các phương pháp tinh chỉnh khác:
1. **Tiếp tục huấn luyện trong miền mới**: Cập nhật toàn bộ tham số của mô hình dựa trên dữ liệu từ miền mới.
2. **Tinh chỉnh tiết kiệm tham số (Parameter-efficient)**: Chỉ tinh chỉnh các tham số mới, nhỏ để thích ứng với miền mới (ví dụ: kỹ thuật LoRA).
3. **Tinh chỉnh theo nhiệm vụ**: Thêm đầu ra chuyên biệt và cập nhật tính năng cho nhiệm vụ đặc thù (ví dụ: phân loại, gán nhãn chuỗi).

Instruction tuning chỉ yêu cầu một phần chi phí nhỏ so với huấn luyện mô hình nền ban đầu, thường chỉ qua vài lượt trên tập dữ liệu hàng nghìn câu lệnh.

***Instructions as Training Data***

Nội dung đề cập đến khái niệm **instruction tuning** trong LLM, sử dụng các mô tả ngôn ngữ tự nhiên về nhiệm vụ kèm theo ví dụ minh họa có nhãn. Phương pháp này giúp mô hình thực hiện đa dạng các nhiệm vụ như trả lời câu hỏi, dịch ngôn ngữ, hay tóm tắt văn bản. Các tập dữ liệu instruction tuning lớn đã được phát triển, bao gồm Aya với 503 triệu câu lệnh từ 114 ngôn ngữ, SuperNatural Instructions với 12 triệu ví dụ, FLAN với 15 triệu ví dụ, và OPT-IML với 18 triệu ví dụ.

Có bốn cách chính để tạo dữ liệu instruction tuning:
1. **Viết thủ công**: Như tập dữ liệu Aya, với 204.000 cặp chỉ dẫn/phản hồi do 3.000 người đóng góp viết tay, nhằm cải thiện hiệu suất của LLM trên nhiều ngôn ngữ.
2. **Sử dụng dữ liệu đã có**: Tận dụng các tập dữ liệu có sẵn cho nhiều nhiệm vụ ngôn ngữ tự nhiên, như SQuAD, và chuyển đổi chúng thành các cặp chỉ dẫn/đầu vào-đầu ra bằng mẫu (template).
3. **Dùng hướng dẫn chú thích**: Tận dụng hướng dẫn chi tiết cho nhiệm vụ chú thích để tạo dữ liệu huấn luyện qua các mẫu câu lệnh.
4. **Sử dụng LLM để tạo dữ liệu**: Dùng LLM để tạo ra các câu hỏi biến thể và câu trả lời an toàn cho các câu hỏi nguy hại, nhằm giúp mô hình học cách phản hồi an toàn hơn.

Phương pháp này đã chứng minh hiệu quả trong việc giảm thiểu khả năng mô hình tạo ra nội dung độc hại.

***Evaluation of Instruction-Tuned Models***

Mục tiêu của **instruction tuning** không phải để học một nhiệm vụ cụ thể, **mà là để học cách tuân thủ các chỉ dẫn một cách tổng quát**. Vì vậy, khi **đánh giá các phương pháp instruction tuning, ta cần xem xét khả năng của mô hình được huấn luyện theo chỉ dẫn trong việc thực hiện các nhiệm vụ mới mà nó chưa được cung cấp chỉ dẫn rõ ràng.**

Cách đánh giá phổ biến là sử dụng phương pháp loại bỏ từng nhiệm vụ — huấn luyện mô hình với một tập hợp lớn các nhiệm vụ và sau đó đánh giá nó trên một nhiệm vụ bị loại bỏ khỏi tập huấn luyện. Tuy nhiên, với số lượng lớn nhiệm vụ trong các tập dữ liệu instruction tuning (ví dụ: 1600 nhiệm vụ trong **SuperNatural Instructions**), nhiều nhiệm vụ có sự chồng chéo; SuperNatural Instructions bao gồm 25 tập dữ liệu về suy diễn văn bản! Rõ ràng, việc kiểm tra trên một tập dữ liệu suy diễn bị loại bỏ trong khi giữ các tập dữ liệu suy diễn khác trong tập huấn luyện sẽ không đánh giá đúng khả năng của mô hình trên suy diễn như một nhiệm vụ mới.

Để giải quyết vấn đề này, các tập dữ liệu instruction tuning lớn được phân thành các nhóm dựa trên sự tương đồng của nhiệm vụ. Phương pháp loại bỏ từng nhiệm vụ sẽ được áp dụng ở mức độ nhóm. Nghĩa là, để đánh giá hiệu suất của mô hình trên phân tích cảm xúc, tất cả các tập dữ liệu về phân tích cảm xúc sẽ bị loại bỏ khỏi tập huấn luyện và chỉ được dùng cho kiểm tra. Điều này còn có lợi ích là cho phép sử dụng một thước đo nhất quán phù hợp cho đánh giá các nhiệm vụ bị loại bỏ. **SuperNatural Instructions** (Wang et al., 2022), chẳng hạn, có 76 nhóm nhiệm vụ trong tổng số 1600 tập dữ liệu của bộ sưu tập này.

Để đánh giá các mô hình instruction-tuning, **SuperNatural Instructions** (Wang et al., 2022) áp dụng phương pháp đánh giá bằng cách phân chia các nhiệm vụ thành từng cụm (clusters) dựa trên sự tương đồng về loại nhiệm vụ. Các cụm nhiệm vụ này được xây dựng để đảm bảo rằng mỗi cụm đại diện cho một loại nhiệm vụ nhất định và bao gồm nhiều tập dữ liệu về nhiệm vụ đó.

**Quy trình đánh giá** như sau:

1. **Phân cụm nhiệm vụ**: Bộ dữ liệu SuperNatural Instructions bao gồm 1600 tập dữ liệu, được chia thành 76 cụm nhiệm vụ khác nhau. Mỗi cụm đại diện cho một loại nhiệm vụ cụ thể, như suy diễn văn bản, phân tích cảm xúc, dịch ngôn ngữ, v.v.

2. **Loại bỏ từng cụm khỏi huấn luyện**: Thay vì loại bỏ từng nhiệm vụ riêng lẻ, SuperNatural Instructions loại bỏ toàn bộ cụm nhiệm vụ khỏi tập huấn luyện khi đánh giá trên nhiệm vụ đó. Ví dụ, để đánh giá khả năng thực hiện của mô hình trên phân tích cảm xúc, tất cả các tập dữ liệu về phân tích cảm xúc sẽ được loại khỏi tập huấn luyện và chỉ được sử dụng trong quá trình kiểm tra.

3. **Đánh giá trên nhiệm vụ bị loại bỏ**: Sau khi mô hình được huấn luyện trên các cụm nhiệm vụ còn lại, nó sẽ được kiểm tra trên các nhiệm vụ trong cụm đã loại bỏ. Cách tiếp cận này giúp kiểm tra xem mô hình có thể thực hiện tốt các nhiệm vụ mới (tức là các nhiệm vụ chưa từng gặp trong huấn luyện) dựa trên khả năng tổng quát hóa các hướng dẫn đã học trước đó.

4. **Sử dụng thước đo phù hợp theo loại nhiệm vụ**: Vì mỗi cụm đại diện cho một loại nhiệm vụ cụ thể, các thước đo đánh giá phù hợp với từng nhiệm vụ có thể được áp dụng. Ví dụ, với các nhiệm vụ suy diễn văn bản, có thể sử dụng độ chính xác (accuracy), trong khi với các nhiệm vụ tóm tắt có thể dùng điểm số ROUGE.

Cách đánh giá này giúp đảm bảo rằng mô hình không chỉ học các mẫu trong dữ liệu huấn luyện mà còn có thể tổng quát hóa và thực hiện tốt trên các nhiệm vụ chưa được huấn luyện trực tiếp, từ đó đo lường khả năng tuân thủ hướng dẫn của mô hình đối với các nhiệm vụ mới.

### Chain-of-Thought Prompting

**chain-of-thought**

Có nhiều kỹ thuật sử dụng lời nhắc (prompts) để cải thiện hiệu suất của các mô hình ngôn ngữ trên nhiều nhiệm vụ. Dưới đây, chúng tôi mô tả một kỹ thuật được gọi là **chuỗi suy nghĩ** (chain-of-thought prompting).

Mục tiêu của chuỗi suy nghĩ là cải thiện hiệu suất trên các nhiệm vụ cần suy luận phức tạp mà các mô hình ngôn ngữ thường thất bại. Ý tưởng chính là con người giải quyết các nhiệm vụ này bằng cách chia nhỏ chúng thành các bước, do đó, chúng ta muốn có ngôn ngữ trong lời nhắc để khuyến khích các mô hình ngôn ngữ phân chia nhiệm vụ theo cách tương tự.

Kỹ thuật này thực tế rất đơn giản: mỗi ví dụ trong lời nhắc few-shot sẽ được bổ sung thêm một số văn bản giải thích các bước suy luận. Mục tiêu là làm cho mô hình ngôn ngữ tạo ra các bước suy luận tương tự cho vấn đề đang được giải quyết, và kết quả của các bước suy luận đó sẽ giúp hệ thống đưa ra câu trả lời đúng.

Nhiều nghiên cứu đã chỉ ra rằng việc bổ sung các bước suy luận vào các ví dụ theo cách này khiến các mô hình ngôn ngữ có nhiều khả năng đưa ra câu trả lời đúng cho các nhiệm vụ suy luận phức tạp (Wei và cộng sự, 2022; Suzgun và cộng sự, 2023b). ### Automatic Prompt Optimization

**Automatic Prompt Optimization**

Với một lời nhắc cho một nhiệm vụ (do con người hoặc máy tính tạo ra), các phương pháp tối ưu hóa lời nhắc tìm kiếm các lời nhắc có hiệu suất cải thiện. Hầu hết các cách tiếp cận này có thể được xem như một dạng tìm kiếm cải tiến lặp lại (Russell và Norvig, 2002) thông qua không gian các lời nhắc có thể để tối ưu hóa hiệu suất cho một nhiệm vụ. Do đó, các cách tiếp cận này đều có chung các thành phần sau:

- **Trạng thái bắt đầu** – Một hoặc nhiều lời nhắc ban đầu được tạo bởi con người hoặc máy tính phù hợp cho một nhiệm vụ nào đó.
- **Thước đo đánh giá** – Một phương pháp để đánh giá mức độ hiệu quả của một lời nhắc trong việc thực hiện nhiệm vụ.
- **Phương pháp mở rộng** – Một phương pháp để tạo ra các biến thể của một lời nhắc.

Do sự biến đổi rất lớn trong cách các lời nhắc cho một nhiệm vụ cụ thể có thể được thể hiện bằng ngôn ngữ, các phương pháp tìm kiếm phải được giới hạn trong một không gian hợp lý. **Beam search** là một phương pháp được sử dụng rộng rãi, kết hợp tìm kiếm theo chiều rộng với một hàng đợi ưu tiên có độ rộng cố định nhằm tập trung nỗ lực tìm kiếm vào các biến thể đạt hiệu suất tốt nhất. Hình 12.11 phác thảo phương pháp chung đằng sau hầu hết các phương pháp tối ưu hóa lời nhắc hiện nay.

Bắt đầu với một hoặc nhiều lời nhắc ứng viên ban đầu, thuật toán tạo ra các biến thể và thêm chúng vào danh sách các lời nhắc cần xem xét. Các lời nhắc này sau đó sẽ được thêm có chọn lọc vào danh sách hoạt động dựa trên việc điểm số của chúng có nằm trong tập ứng viên hàng đầu hay không. Độ rộng của beam bằng 1 dẫn đến một tìm kiếm tham lam tập trung, trong khi độ rộng beam vô hạn dẫn đến tìm kiếm theo chiều rộng toàn diện. Mục tiêu là tiếp tục tìm kiếm các lời nhắc được cải thiện tùy theo tài nguyên tính toán sẵn có.

Các tìm kiếm cải tiến lặp lại thường sử dụng kết hợp giữa một số lượng vòng lặp cố định và tiêu chí dừng khi không có cải thiện sau một khoảng thời gian nhất định. Tiêu chí này tương đương với việc dừng sớm khi không đạt được cải thiện trong quá trình huấn luyện mạng nơ-ron sâu.

````python
def prompt_optimization(prompts, width):
    # Khởi tạo tập hợp các prompt ứng cử viên ban đầu
    active = prompts
    done = False

    while not done:
        frontier = []  # Khởi tạo tập frontier rỗng
        for c in active:
            children = expand(c)  # Mở rộng từng prompt ứng cử viên
            for child in children:
                # Áp dụng hàm add_to_beam để thêm các prompt vào frontier
                frontier = add_to_beam(child, frontier, width)

        # Cập nhật trạng thái active với frontier
        active = frontier

    return active

def add_to_beam(state, agenda, width):
    # Nếu độ dài của agenda chưa đạt đến width
    if len(agenda) < width:
        agenda.append(state)  # Thêm trạng thái mới vào agenda
    else:
        # Tính điểm của trạng thái hiện tại và điểm tệ nhất trong agenda
        if score(state) > score(worst_of(agenda)):
            # Xóa trạng thái có điểm thấp nhất trong agenda
            agenda.remove(worst_of(agenda))
            # Thêm trạng thái mới vào agenda
            agenda.append(state)

    return agenda

# Các hàm bổ trợ cần thiết (ví dụ)
def expand(c):
    # Hàm mở rộng các prompt, trả về danh sách các prompt con
    # Chỉ là ví dụ, bạn cần điều chỉnh cho phù hợp với bài toán cụ thể
    return [c + "1", c + "2", c + "3"]

def score(state):
    # Hàm tính điểm cho một trạng thái cụ thể
    # Thay thế bằng logic tính điểm cụ thể của bạn
    return len(state)  # Ví dụ: điểm dựa trên độ dài chuỗi

def worst_of(agenda):
    # Trả về trạng thái có điểm thấp nhất trong agenda
    return min(agenda, key=score)
````

````python
prompts = ["A", "B", "C"]
width = 2
optimized_prompts = prompt_optimization(prompts, width)
print("Optimized Prompts:", optimized_prompts)
````

**Đánh Giá Ứng Viên (Candidate Scoring)**  
Phương pháp đánh giá ứng viên nhằm xác định hiệu suất tiềm năng của các lời nhắc (prompts), để nhận diện các hướng đi triển vọng và loại bỏ những lời nhắc có khả năng không hiệu quả. Vì đánh giá ứng viên được thực hiện trong vòng lặp nội bộ của quá trình tìm kiếm, chi phí tính toán để đánh giá là rất quan trọng.

- **Độ chính xác khi thực thi (Execution Accuracy)**  
  Với quyền truy cập vào dữ liệu đào tạo có nhãn, các lời nhắc ứng viên có thể được đánh giá dựa trên độ chính xác khi thực thi (Honovich et al., 2023). Trong phương pháp này, lời nhắc ứng viên được kết hợp với các đầu vào từ tập dữ liệu đào tạo và truyền vào một mô hình ngôn ngữ lớn (LLM) để giải mã. Đầu ra của LLM sẽ được đánh giá so với nhãn của tập dữ liệu dựa trên một thước đo phù hợp cho nhiệm vụ. Đối với các nhiệm vụ dựa trên phân loại, đây là phép đo mất mát 0/1 – có bao nhiêu ví dụ được dán nhãn đúng với lời nhắc đã cho. Các ứng dụng tạo sinh như tóm tắt hoặc dịch thuật sử dụng các thước đo điểm tương đồng chuyên biệt cho nhiệm vụ như BERTScore, Bleu (Papineni et al., 2002) hoặc ROUGE (Lin, 2004). Do chi phí tính toán để gọi LLM, việc đánh giá từng lời nhắc ứng viên so với toàn bộ tập dữ liệu đào tạo sẽ không khả thi. Thay vào đó, hiệu suất của lời nhắc được ước tính từ một mẫu nhỏ của dữ liệu đào tạo (Pryzant et al., 2023).

**Mở Rộng Lời Nhắc (Prompt Expansion)**  
Phương pháp mở rộng lời nhắc tạo ra các biến thể của một lời nhắc đã cho nhằm tạo ra một tập hợp các lời nhắc mở rộng có thể cải thiện hiệu suất so với bản gốc. Một phương pháp phổ biến là sử dụng các mô hình ngôn ngữ để tạo các phiên bản diễn giải lại. Ví dụ, Zhou et al. (2023) sử dụng một meta-prompt sau để yêu cầu biến thể từ lời nhắc gốc:
- **Yêu cầu Biến Thể của Lời Nhắc (Prompting for a Variant)**  
  Tạo một biến thể của hướng dẫn sau nhưng vẫn giữ nguyên ý nghĩa ngữ nghĩa.  
  **Đầu vào**: HƯỚNG DẪN  
  **Đầu ra**: HOÀN TẤT  

Một biến thể của phương pháp này là cắt ngắn lời nhắc hiện tại ở các vị trí ngẫu nhiên, tạo ra một tập hợp các tiền tố của lời nhắc. Sau đó, LLM diễn giải sẽ được yêu cầu tiếp tục các tiền tố này để tạo ra một lời nhắc hoàn chỉnh.

Đây là một ví dụ về **tìm kiếm không có định hướng (uninformed search)**, nghĩa là bước mở rộng ứng viên không được định hướng để tạo ra các ứng viên tốt hơn; các ứng viên được tạo ra mà không cần quan tâm đến chất lượng. Vai trò của hàng đợi ưu tiên là nâng cao các ứng viên đã được cải thiện khi chúng được phát hiện. Ngược lại, Prasad et al. (2023) áp dụng một kỹ thuật mở rộng ứng viên nhằm tạo ra các lời nhắc vượt trội trong quá trình mở rộng. Trong cách tiếp cận này, ứng viên hiện tại đầu tiên được áp dụng cho một mẫu các ví dụ đào tạo bằng phương pháp độ chính xác khi thực thi. Hiệu suất của lời nhắc trên các ví dụ này sau đó hướng dẫn quá trình mở rộng. Cụ thể, các ví dụ không chính xác được sử dụng để đánh giá lại lời nhắc gốc – với đánh giá đóng vai trò là một gradient trong quá trình tìm kiếm.

Phương pháp bao gồm các bước sau:
1. Chạy lời nhắc trên một mẫu của các ví dụ đào tạo.
2. Xác định các ví dụ mà lời nhắc không thành công.
3. Yêu cầu LLM đưa ra đánh giá về lời nhắc dựa trên các ví dụ thất bại.
4. Cung cấp đánh giá này cho LLM và yêu cầu nó tạo ra các lời nhắc cải tiến.

Với một lời nhắc và một tập hợp các ví dụ thất bại, Prasad et al. (2023) sử dụng mẫu sau để yêu cầu đánh giá từ một LLM cho nhiệm vụ phân loại.

- **Đánh Giá Lời Nhắc (Critiquing Prompt)**  
  Tôi đang cố gắng viết một lời nhắc phân loại không dựa trên ngữ cảnh ban đầu.  
  Lời nhắc hiện tại của tôi là: **prompt**  
  Nhưng lời nhắc này mắc lỗi với các ví dụ sau: **error string**  
  Hãy cung cấp **num feedbacks** lý do tại sao lời nhắc có thể đã mắc lỗi với các ví dụ này.

Phản hồi của mô hình sau đó được kết hợp với một mẫu thứ hai để yêu cầu LLM tạo ra các lời nhắc cải tiến.

- **Yêu Cầu Cải Tiến Lời Nhắc (Prompt Improvement Prompt)**  
  Tôi đang cố gắng viết một lời nhắc phân loại không dựa trên ngữ cảnh ban đầu. Lời nhắc hiện tại của tôi là: **prompt**  
  Nhưng nó mắc lỗi với các ví dụ sau: **error**  
  Dựa trên các ví dụ này, vấn đề với lời nhắc này là **gradient**.  
  Dựa trên thông tin trên, tôi đã viết các bước **per gradient** khác nhau để tạo các lời nhắc mới cải tiến.

### ROUGE

ROUGE (Nghiên cứu theo định hướng thu hồi để đánh giá Gisting), là một tập hợp các số liệu và gói phần mềm được thiết kế đặc biệt để đánh giá tóm tắt tự động, nhưng cũng có thể được sử dụng để dịch máy. ROUGE (Recall-Oriented Understudy for Gisting Evaluation), is a set of metrics and a software package specifically designed for evaluating automatic summarization, but that can be also used for machine translation. The metrics compare an automatically produced summary or translation against reference (high-quality and human-produced) summaries or translations.

- Các phương pháp như chuỗi suy luận (chain-of-thought) có thể được sử dụng để tạo ra các lời nhắc giúp các mô hình ngôn ngữ xử lý các vấn đề suy luận phức tạp.
- Các mô hình ngôn ngữ đã được tiền huấn luyện có thể được điều chỉnh để hoạt động theo các cách mong muốn thông qua sự căn chỉnh mô hình (model alignment).
- Một phương pháp để căn chỉnh mô hình là tinh chỉnh theo hướng dẫn (instruction tuning), trong đó mô hình được tinh chỉnh (sử dụng mục tiêu dự đoán từ tiếp theo của mô hình ngôn ngữ) trên một tập dữ liệu các hướng dẫn cùng với các phản hồi đúng. Các tập dữ liệu tinh chỉnh theo hướng dẫn thường được tạo ra bằng cách sử dụng lại các tập dữ liệu NLP tiêu chuẩn cho các nhiệm vụ như trả lời câu hỏi hoặc dịch máy.
